## Dataset exploration

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
DATA_PATH_IMG = "../Chest-X-Ray/image"
DATA_PATH_MSK = "../Chest-X-Ray/mask"

patient_folders_imgs = sorted(glob(os.path.join(DATA_PATH_IMG, '[0-9]*')))
patient_folders_mks = sorted(glob(os.path.join(DATA_PATH_MSK, '[0-9]*')))

print(f"Total number of patients: {len(patient_folders_imgs)}")
print(f"\nFirst 5 patient folders:")
for folder in patient_folders_imgs[:5]:
    print(os.path.basename(folder))
```

```python
idx = 100

sample_patient_img = patient_folders_imgs[idx]
sample_patient_msk = patient_folders_mks[idx]

img = cv2.imread(sample_patient_img)
mask = cv2.imread(sample_patient_msk, cv2.IMREAD_GRAYSCALE)

# Display
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(mask, cmap='gray')
axes[1].set_title('Mask')
axes[1].axis('off')

axes[2].imshow(img)
axes[2].imshow(mask, cmap='Reds', alpha=0.5)  # overlay
axes[2].set_title('Overlay')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
